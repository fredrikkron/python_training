### Matplotlob

```python
import matplotlib.pyplot as plt

f = lambda x: 3 * x - 2

def g(x):
    return x**2 -2

x = tuple(range(-5,5))

y1 = [f(i) for i in x]
y2 = [g(i) for i in x]

plt.plot(x, y1)
plt.plot(x, y2)
plt.title("plotting 2 functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["$f(x)=3x-2$", "$g(x) = x^2 - 2$"])
```

```python
import numpy as np

x = np.linspace(-5,5)
x
```

```python
x.shape
```

```python
f = lambda x: x**2 + 1
g = lambda x: np.sin(x) + 5

# OOP approach
fig, ax = plt.subplots()
ax.plot(x, f(x), label = "$f(x) = x^2 + 1$")
ax.plot(x, g(x), label = "$g(x) = \sin x + 5$")
ax.set(title = "Two graphs", xlabel="x", ylabel="y")
ax.legend()
```
